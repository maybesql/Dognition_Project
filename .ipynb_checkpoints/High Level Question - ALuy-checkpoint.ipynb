### Import all the necessary libraries:

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Import the dognition_data_aggregated_by_dogid dataset into pandas data frame:

```python
df = pd.read_excel('data/dognition_data_aggregated_by_dogid.xlsx')
```

```python
df_DogFeats = df.loc[:,['Dog ID','Total Tests Completed','Mean ITI (days)','Mean ITI (minutes)','Median ITI (days)','Median ITI (minutes)','Time diff between first and last game (days)','Time diff between first and last game (minutes)','Gender','Birthday','Breed','Breed_Type','Breed_Group','Weight','Dog_Fixed','DNA_Tested','Dimension',]]
df_DogFeats.head()
```

### How does Dog_Fixed correlated to the number of tests completed?

```python
# group the total test completed by Dog_Fixed
Dog_Fixed_df = df_DogFeats.groupby(df_DogFeats['Dog_Fixed'])['Total Tests Completed'].sum()
Dog_Fixed_df
```

```python
#Pie chart for sum of tests completed Fixed Dog column
import matplotlib.pyplot as plt

labels = Dog_Fixed_df.index
sizes = Dog_Fixed_df.values


fig1, ax1 = plt.subplots(figsize = (14,8))
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  
plt.legend()
plt.title('Pie chart for sum of tests completed Fixed Dog column', fontsize = 15)
plt.show()
```

#### From the pie chart above we can see that those dogs who have fixed status completed the most tests.

### How do the Weight correlated to the number of tests completed?

```python
plt.figure(figsize = (14,9))
plt.scatter(df_DogFeats['Total Tests Completed'] ,df_DogFeats['Weight'])
plt.xlabel("Weight",fontsize=15)  
plt.ylabel("Sum of TOtal Tests",fontsize=15)
plt.title('A Bar graph showing Sum of TOtal Tests for Weight column',fontsize=15)
plt.show()
```

#### From the above scatter plot we can see that as the number of tests inceases above 20, the dogs with heavy Weights could complete lesser tests.

### How do the Membership_Type correlated to the number of tests completed?

```python
# Replace CA with 0 and also filled the missing values 
df['Membership_Type'] = df['Membership_Type'].replace('CA', 0)
df['Membership_Type'].fillna(df['Membership_Type'].mode()[0], inplace = True)
```

```python
# group the total test completed by Membership_Type
Membership_Type_df = df.groupby(df['Membership_Type'])['Total Tests Completed'].sum()
Membership_Type_df
```

```python
ax = Membership_Type_df.plot(figsize=(14,9) ,fontsize=15, linewidth = 3)
ax.set_xlabel("Membership_Type",fontsize=15)  
ax.set_ylabel("Sum of TOtal Tests",fontsize=15)
ax.set_title('A Bar graph showing Sum of TOtal Tests for Membership_Type column',fontsize=15)
plt.show()
```

#### From the line chart above we can see that Membership_Type 1  completed the most tests.

### How do the owner feature Subscribed correlated to the number of tests completed?

```python
ax = df['Subscribed'].value_counts().plot(kind='bar',figsize=(14,9), width=0.40 ,fontsize=15, color = ['red','yellow'])
ax.set_xlabel("Subscribed",fontsize=15)  
ax.set_ylabel("Sum of Total Tests",fontsize=15)
ax.set_title('A Bar graph showing Sum of Total Tests for Subscribed column',fontsize=15)
plt.show()
```

##### We can above in the bar chart that the owners who had bought a paid Subscription their dogs completed the most of the tests.
